```python
import pandas as pd
```

```python
dataset = [['<=30','high','false','fair','no'],
          ['<=30','high','false','excellent','no'],
          ['31...40','high','false','fair','yes'],
          ['>40','medium','false','fair','yes'],
          ['>40','low','true','fair','yes'],
          ['>40','low','true','excellent','no'],
          ['31...40','low','true','excellent','yes'],
          ['<=30','medium','false','fair','no'],
          ['<=30','low','true','fair','yes'],
          ['>40','medium','true','fair','yes'],
          ['<=30','medium','true','excellent','yes'],
          ['31...40','medium','false','excellent','yes'],
          ['31...40','high','true','fair','yes'],
          ['>40','medium','false','excellent','no']]
```

```python
df = pd.DataFrame(dataset, columns = ['age','income','student', 'credit_rating', 'buys_computer'])
```

```python
df
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
x=apriori(df, min_support=0.1, use_colnames=True)
```

```python
temp = x
temp
```

```python
temp['length']=temp['itemsets'].apply(lambda x:len(x))
temp
```

```python
temp = temp[temp['length']>=2]
```

```python
temp
```

```python
from mlxtend.frequent_patterns import association_rules
result = association_rules(temp,metric='confidence',min_threshold=0.6)
result
```

```python
x1 = result.loc[result['consequents']== {'yes'}]
x2 = result.loc[result['consequents']== {'no'}]
x3 = pd.concat([x1,x2])
```

```python
temp1 = x3.loc[x3['antecedents'].str.contains('<=30',regex=False)]
temp2 = x3.loc[x3['antecedents'].str.contains('medium',regex=False)]
temp3 = x3.loc[x3['antecedents'].str.contains('true',regex=False)]
temp4 = x3.loc[x3['antecedents'].str.contains('fair',regex=False)]
pd.concat([temp1,temp2,temp3,temp4]).drop_duplicates().reset_index(drop=True)
```

### Only index with 0,1,2,7,10,11,13,17,21 satisfied all test input.

### The highest confident rules predicts class = Yes.
### Thus predict Yes.
